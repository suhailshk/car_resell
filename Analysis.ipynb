```python
fatal = data1.iloc[0].sort_values(ascending=False).head(5)
```

```python
explode = (0.1, 0.0, 0.2, 0.3, 0.1)
colors = ( "orange", "cyan", "brown",
          "grey", "indigo")

wp = { 'linewidth' : 1, 'edgecolor' : "black" }


fatal.plot(kind='pie', autopct = '%1.1f%%',explode = explode,
                                  shadow = True,
                                  colors = colors,
                                  startangle = 90,
                                  wedgeprops = wp,
                                  textprops = dict(color ="magenta"))
```
